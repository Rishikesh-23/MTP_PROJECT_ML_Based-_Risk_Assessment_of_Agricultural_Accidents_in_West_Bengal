```python
import matplotlib.pyplot as plt
import numpy as np

# Data to use for comparison
categories = ['GDP Contribution', 'Workforce in Agriculture', 'Land Use for Agriculture']
india_values = [18.3, 42.3, 60]  # Assume 60% of India's land is agricultural
world_values = [5, 27, 37]       # Assume 37% of global land is agricultural

# Setting custom colors
india_color = '#1f77b4'
world_color = '#ff7f0e'

# 1. Side-by-Side Bar Plot for Comparison
plt.figure(figsize=(10, 6))
x = np.arange(len(categories))
width = 0.35

# Plot bars
plt.bar(x - width/2, india_values, width, label='India', color=india_color)
plt.bar(x + width/2, world_values, width, label='World', color=world_color)

# Labeling
plt.ylabel('Percentage (%)')
plt.title('Comparison of Agricultural Metrics: India vs World')
plt.xticks(ticks=x, labels=categories)
plt.legend()

# Adding values on top of bars
for i, v in enumerate(india_values):
    plt.text(x[i] - width/2, v + 1, f"{v}%", ha='center')
for i, v in enumerate(world_values):
    plt.text(x[i] + width/2, v + 1, f"{v}%", ha='center')

plt.tight_layout()
plt.show()

# 2. Horizontal Bar Plot for Land Use Comparison
plt.figure(figsize=(8, 4))
categories_land = ['India', 'World']
land_use_values = [60, 37]

plt.barh(categories_land, land_use_values, color=[india_color, world_color])
plt.xlabel('Percentage of Land Used for Agriculture')
plt.title('Land Use for Agriculture: India vs World')
for index, value in enumerate(land_use_values):
    plt.text(value, index, f"{value}%")
plt.show()

# 3. Combined Plot with Multiple Metrics
plt.figure(figsize=(10, 6))
metrics = ['GDP Contribution (%)', 'Workforce (%)', 'Land Use (%)']
india_metrics = india_values
world_metrics = world_values

x_axis = np.arange(len(metrics))

plt.plot(x_axis, india_metrics, 'o-', label='India', color=india_color, linewidth=2)
plt.plot(x_axis, world_metrics, 's--', label='World', color=world_color, linewidth=2)

plt.xticks(ticks=x_axis, labels=metrics)
plt.ylabel('Percentage')
plt.title('India vs World: Agricultural Metrics Comparison')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for India vs World comparison on agricultural fatalities and economic impact
categories = ['Annual Fatalities', 'Economic Loss ($)']
india_data = [10000, 400e6]  # India: 10,000 fatalities, $400 million in losses
world_data = [170000, 1e12]  # World: 170,000 fatalities, $1 trillion in losses

# Convert data for plotting
X = np.arange(len(categories))
width = 0.35  # width of the bars

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars_india = ax.bar(X - width/2, india_data, width, label='India', color='#FF7F0E')
bars_world = ax.bar(X + width/2, world_data, width, label='World', color='#2E8B57')

# Labels, title, and custom x-axis tick labels
ax.set_xlabel('Category', fontsize=14, fontweight='bold')
ax.set_ylabel('Value', fontsize=14, fontweight='bold')
ax.set_title('Agricultural Fatalities and Economic Impact: India vs World', fontsize=16, fontweight='bold')
ax.set_xticks(X)
ax.set_xticklabels(categories)

# Display values on top of bars
for bar in bars_india:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{int(height):,}', ha='center', va='bottom', fontsize=12)
for bar in bars_world:
    height = bar.get_height()
    label = f'${height/1e9:.1f} B' if height > 1e9 else f'{int(height):,}'
    ax.text(bar.get_x() + bar.get_width() / 2, height, label, ha='center', va='bottom', fontsize=12)

# Legend and grid
ax.legend(title="Region", fontsize=12, loc='upper left')
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define percentage data (based on India vs World comparison)
categories = ['Fatalities', 'Economic Impact']
india_percent = [5.9, 40]  # Percentages for India (e.g., 5.9% of global fatalities, 40% of economic impact)
world_percent = [94.1, 60]  # World as remaining percentage

# Stack bars
fig, ax = plt.subplots(figsize=(10, 6))
X = np.arange(len(categories))
ax.bar(X, india_percent, width=0.5, label='India', color='#FF7F0E')
ax.bar(X, world_percent, width=0.5, bottom=india_percent, label='World', color='#2E8B57')

# Customize labels and title
ax.set_xticks(X)
ax.set_xticklabels(categories)
ax.set_xlabel('Category', fontsize=14, fontweight='bold')
ax.set_ylabel('Percentage (%)', fontsize=14, fontweight='bold')
ax.set_title('India vs World: Agricultural Fatalities & Economic Impact in Percentage', fontsize=16, fontweight='bold')

# Show values on the bars
for i, (india, world) in enumerate(zip(india_percent, world_percent)):
    ax.text(i, india / 2, f'{india}%', ha='center', va='center', color='white', fontsize=12)
    ax.text(i, india + (world / 2), f'{world}%', ha='center', va='center', color='white', fontsize=12)

ax.legend(title="Region")
plt.tight_layout()
plt.show()
```

```python
# Horizontal bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(categories, india_percent, color='#FF7F0E', label='India')
ax.barh(categories, world_percent, left=india_percent, color='#2E8B57', label='World')

# Label bars with percentages
for i, (india, world) in enumerate(zip(india_percent, world_percent)):
    ax.text(india / 2, i, f'{india}%', va='center', ha='center', color='white', fontsize=12)
    ax.text(india + (world / 2), i, f'{world}%', va='center', ha='center', color='white', fontsize=12)

ax.set_xlabel('Percentage (%)', fontsize=14, fontweight='bold')
ax.set_title('India vs World: Agricultural Fatalities & Economic Impact in Percentage', fontsize=16, fontweight='bold')
ax.legend(title="Region")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Colors for India and World segments
colors = ['#FF7F0E', '#2E8B57']

# Pie chart for Fatalities
axes[0].pie([india_percent[0], world_percent[0]], labels=['India', 'World'], autopct='%1.1f%%', startangle=90, colors=colors)
axes[0].set_title('Fatalities', fontsize=14, fontweight='bold')

# Pie chart for Economic Impact
axes[1].pie([india_percent[1], world_percent[1]], labels=['India', 'World'], autopct='%1.1f%%', startangle=90, colors=colors)
axes[1].set_title('Economic Impact', fontsize=14, fontweight='bold')

plt.suptitle('India vs World: Percentage Comparison', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
